```python
train_mean = np.mean(train_features, axis=0)
train_std = np.std(train_features, axis=0)
train_features = (train_features - train_mean) / train_std
```

Build A Deep Learning Model <br>

The model would have a hidden layer of 20 neurons

**Loss Function** - Loss function needs to be minimised to find the parameters of the model during training <br>
We are using loss function as mean squared error <br>
**Optimizer** - Adam - extended version of stochastic gradient descent optimizer
**Metrics** - Metrics are similar to loss function except that they are not used for training the data <br>
mse stands for mean squared error <br>
mae stands for mean absolute error <br>






```python
def build_model():
    model = keras.Sequential([
        Dense(20, activation=tf.nn.relu, input_shape=[len(train_features[0])]),
        Dense(10, activation=tf.nn.relu),
        Dense(1)
    ])

    model.compile(optimizer=tf.optimizers.Adam(), 
                  loss='mse',
                 metrics=['mse','mae'])
    #model.compile('sgd', metrics=['mean_absolute_error'],loss='mse')
    return model
```

```python
def build_model():
    model = keras.Sequential([
        Dense(20, activation=tf.nn.relu, input_shape=[len(train_features[0])]),
        Dense(1)
    ])

    model.compile(optimizer=tf.optimizers.Adam(), 
                  loss='mse',
                 metrics=['mse','mae'])
    #model.compile('sgd', metrics=['mean_absolute_error'],loss='mse')
    return model
```

```python
model = build_model()
model.summary()
```

First 90 of the data points would be training data set and the next 10 percent of the data points would be testing data set<br>
One epoch means that every data point in the training set got an opportunity to update the internal weights of the model<br>
Verbose = 0 means nothing displayed in colab


```python
model_fit = model.fit(train_features, train_labels, epochs=1000, verbose=2, validation_split = 0.2)
```

```python
model_fit_hist_df = pd.DataFrame(model_fit.history)
#print(model_fit_hist_df)
model_fit_hist_df['epoch']= model_fit.epoch
#hist['epoch'] = model_fit.epoch
print(model_fit_hist_df)



```

```python
def plot_history():
    plt.figure(figsize=(20,8))
    plt.xlabel('Number of Epochs')
    plt.ylabel('Mean Square Error [Thousand Dollars$^2$]')
    plt.plot(model_fit_hist_df['epoch'], model_fit_hist_df['mse'], label='Training Error')
    plt.plot(model_fit_hist_df['epoch'], model_fit_hist_df['val_mse'], label = 'Validation Error')
    plt.legend()
    plt.ylim([0,50])

plot_history()
```
