centrality measures

```python
# importing libraries
import networkx as nx
import matplotlib.pyplot as plt
my_graph = nx.Graph()
my_graph.add_edge(1, 2)
my_graph.add_edge(1, 4)
my_graph.add_edge(2, 4)
my_graph.add_edge(2, 1)
my_graph.add_edge(3, 2)
nx.draw(my_graph,with_labels=True)
#sorting betweenness centrality for each node and reporting the nodes in the order of decreasing betweenness centrality
sorted(nx.betweenness_centrality(my_graph, normalized=True).items(), key=lambda x:x[1], reverse=True)[0:10]
# the result would be node followed by the betweenness centrality measures

```

```python
sorted(nx.degree_centrality(my_graph).items(), key=lambda x:x[1], reverse=True)[0:10]
```
